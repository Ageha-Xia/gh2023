```python
import torch
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open('../data/eventimage.pkl', 'rb') as f:
    EventImage = pickle.load(f)
    ArrivalTime, ArrivalCount = zip(*EventImage)
    # ArrivalTimeImage = np.stack(list(ArrivalTime))
    # ArrivalCountImage = np.stack(list(ArrivalCount))
```

```python
def plot_image(data, index):
    image = data[index]
    plt.imshow(image, cmap='jet')
    plt.colorbar()
    plt.title(f"Image at Index {index}")
    plt.show()
```

```python
n = 10
plot_image(ArrivalTime, n)
plot_image(ArrivalCount, n)
```
